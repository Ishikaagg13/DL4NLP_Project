```python
import json

# Path to the JSON file
json_file_path = 'D:\JIO Institute\Q4\DL_NLP\project\datasets\qna_transformed.json'

# Open the JSON file
with open(json_file_path, 'r') as file:
    # Read the JSON content
    data = json.load(file)

# Display the JSON content
print(data)
```

```python
for i in range(0,8):
    print(data[i])
```

```python
import pandas as pd
# Create a DataFrame
df = pd.DataFrame(columns=['messages', 'images'])
```

```python
import os

# Path to the folder containing images
folder_path = 'Images'

# Get a list of all files in the folder
files = os.listdir(folder_path)

# Filter out only the image files
image_files = [file for file in files if file.lower().endswith(('png', 'jpg', 'jpeg', 'gif', 'bmp'))]


# Create a list of image paths
image_paths = [os.path.join(folder_path, file) for file in image_files]

    
sorted_image_paths = sorted(image_paths, key=lambda x: int(x.split('\\')[-1].split('_')[0]))

for path in sorted_image_paths:
    print(path)
```

```python
# Iterate over the image paths and add rows to the DataFrame



for i,j in zip( range(0,len(data),8), sorted_image_paths):
    # Create a dictionary representing the row
    clubbed_messages = []

    try:
        clubbed_messages.append(data[i:i+8])
        
    except:
        print("kaam khatam paisa hazam !!!!")
        
    row = {
        'messages': clubbed_messages,
        'images': j
    }
    # Add the row to the DataFrame
    df = df.append(row, ignore_index=True)

# Display the DataFrame
print(df)
```

```python
# Save the DataFrame to an Excel file
df.to_excel('final_data_V1.xlsx', index=False)

```
