```python
import os
import requests
import uuid

def download_image(url, folder_path,i):
    """
    Download an image from the given URL and save it in a folder with a unique ID as the filename.
    
    Args:
        url (str): The URL of the image to download.
        folder_path (str): The path to the folder where the image will be saved.
        
    Returns:
        str: The path to the saved image.
    """
    # Create the folder if it doesn't exist
    os.makedirs(folder_path, exist_ok=True)
    
    # Generate a unique ID for the image
    unique_id = str(uuid.uuid4())[:8]
    
    # Get the file extension from the URL
    file_extension = url.split('.')[-1]
    
    # Construct the file path
    file_path = os.path.join(folder_path, f"{i}_{unique_id}.{file_extension}")
    
    # Download the image
    response = requests.get(url)
    with open(file_path, 'wb') as file:
        file.write(response.content)
    
    return file_path


# Example usage
url = 'https://example.com/image.jpg'
folder_path = 'Images'

j = 1
for i in data:
   
    download_image(i["Image"], folder_path, j)
    j+=1
```
