### image downloader

```python
import json
import os
import requests
import csv
import uuid

# Path to your JSON file
json_file_path = r'D:\JIO Institute\Q4\DL_NLP\project\musuem.json'

# Load the dataset from the JSON file
with open(json_file_path, 'r', encoding='utf-8') as file:
    dataset = json.load(file)

# Create a folder to save the images
os.makedirs('images', exist_ok=True)

# Create a CSV file to store the unique IDs, URLs, and paths
csv_file_path = 'image_metadata.csv'

with open(csv_file_path, 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerow(['img_id', 'img_url', 'img_path'])

    # Download images and save them to the folder
    for item in dataset:
        image_url = item['Image']
        image_name = image_url.split('/')[-1]
        unique_id = str(uuid.uuid4())  # Generate a unique ID for the image
        image_path = os.path.join('images', image_name)
        response = requests.get(image_url)
        with open(image_path, 'wb') as f:
            f.write(response.content)
        csv_writer.writerow([unique_id, image_url, image_path])

print("CSV file created with image metadata.")


```

```python

formatted_dataset = []
for item in dataset:
    title_parts = item["Title"].split("\n")
    
    filtered_title_list = [item for item in title_parts if item != ""]
    
    try:
        formatted_item = {
            "Title": filtered_title_list[0],
            "Code": filtered_title_list[1],
            "Material": filtered_title_list[2],
            "Origin": filtered_title_list[3],
            "Date": filtered_title_list[4],
            "Dimensions": filtered_title_list[5],
            "Description": filtered_title_list[6],
            "Image": item["Image"]
        }
        formatted_dataset.append(formatted_item)
    except IndexError:
        print(item)

# Convert the formatted dataset to JSON format
formatted_dataset_json = json.dumps(formatted_dataset, indent=4)
print(formatted_dataset_json)
with open('formatted_dataset.json', 'w') as file:
    json.dump(formatted_dataset, file, indent=4)

```
